# Partie 3 — Complexité du calcul d'un équilibre de Nash

Le calcul d'un équilibre de Nash est un problème central en théorie algorithmique des jeux. Sa complexité est particulière, car il ne s'agit pas simplement de répondre oui ou non à une question. En effet, le théorème de Nash garantit que tout jeu fini possède au moins un équilibre, éventuellement en stratégies mixtes. Le véritable problème algorithmique est donc le suivant : **comment trouver effectivement un équilibre de Nash ?**

## 1. Un problème de recherche plutôt qu'un problème de décision

Les classes classiques comme **P** et **NP** concernent surtout des problèmes de décision, où la réponse attendue est oui ou non. Pour l'équilibre de Nash, l'existence d'une solution est déjà garantie. On s'intéresse donc à un **problème de recherche totale** : une solution existe toujours, mais elle peut être difficile à calculer.

C'est pourquoi on utilise la classe **PPAD**, qui signifie *Polynomial Parity Arguments on Directed graphs*. Cette classe regroupe des problèmes de recherche pour lesquels l'existence d'une solution est garantie par un argument de parité dans un graphe orienté.

## 2. La classe PPAD et le problème END OF THE LINE

Le problème canonique de PPAD est appelé **END OF THE LINE**. De manière simplifiée, on considère un très grand graphe orienté dans lequel chaque sommet a au plus un prédécesseur et au plus un successeur. On connaît un sommet de départ d'une ligne, et le but est de trouver une autre extrémité de ligne.

L'existence d'une telle extrémité est garantie par un argument de parité : dans un graphe fini, si une ligne commence quelque part, elle doit soit se terminer, soit former une structure qui force l'existence d'une autre extrémité. Le problème n'est donc pas de prouver qu'une solution existe, mais de la trouver efficacement.

Un problème est dit **PPAD-complet** s'il est dans PPAD et si tout problème de PPAD peut être réduit polynomialement à lui. Cela signifie qu'il fait partie des problèmes les plus difficiles de cette classe.

## 3. Pourquoi l'équilibre de Nash appartient à PPAD ?

L'existence d'un équilibre de Nash repose sur un théorème de point fixe, notamment le **théorème du point fixe de Brouwer**. Intuitivement, on peut associer à un jeu une fonction continue qui transforme un profil de stratégies mixtes en un autre profil. Un équilibre de Nash correspond alors à un point fixe de cette fonction.

Le lien avec PPAD vient du fait que la recherche d'un point fixe de Brouwer peut être discrétisée et formulée comme un problème de type END OF THE LINE. On obtient alors la chaîne d'idées suivante :

```text
END OF THE LINE → point fixe discret de Brouwer → équilibre de Nash
```

Ainsi, le calcul d'un équilibre de Nash appartient naturellement à la famille des problèmes dont la solution est garantie par un argument topologique ou combinatoire, mais dont le calcul peut être difficile.

## 4. PPAD-complétude du calcul d'un équilibre de Nash

En 2006, Daskalakis, Goldberg et Papadimitriou ont montré que le calcul d'un équilibre de Nash est **PPAD-complet**. Ce résultat est fondamental : il indique qu'il ne faut pas s'attendre à trouver un algorithme polynomial général pour calculer un équilibre de Nash, sauf si toute la classe PPAD admet des algorithmes polynomiaux.

Autrement dit, si l'on trouvait un algorithme polynomial pour calculer un équilibre de Nash dans le cas général, alors tous les problèmes de PPAD pourraient être résolus efficacement. Cela placerait le calcul de Nash parmi les problèmes de recherche totale les plus faciles de PPAD, ce qui est considéré comme peu probable.

Il faut toutefois distinguer plusieurs cas :

- pour les jeux à deux joueurs, certains problèmes liés aux équilibres exacts ont une structure particulière ;
- pour les jeux à plusieurs joueurs, le calcul et même la représentation exacte des probabilités peuvent devenir plus délicats ;
- les résultats de dureté concernent surtout le calcul général d'un équilibre, notamment dans les jeux représentés de manière explicite.

## 5. Équilibre exact

Un **équilibre de Nash exact** est un profil de stratégies dans lequel aucun joueur ne peut améliorer son gain en changeant seul de stratégie. Dans un jeu fini, cet équilibre peut utiliser des stratégies mixtes, donc des probabilités sur les actions disponibles.

Calculer un équilibre exact signifie déterminer précisément ces probabilités. Cela peut être difficile pour plusieurs raisons :

- les probabilités peuvent être rationnelles ou algébriques ;
- leur représentation exacte peut nécessiter une grande précision ;
- dans le cas général, aucun algorithme polynomial n'est connu.

Le calcul exact est donc important théoriquement, mais il est souvent moins adapté aux situations pratiques, où une petite erreur est acceptable.

## 6. Équilibre ε-approché

Un **équilibre ε-approché** est un profil de stratégies dans lequel aucun joueur ne peut augmenter son gain de plus de ε en déviant seul. Formellement, pour tout joueur, le gain obtenu est à au plus ε du meilleur gain qu'il pourrait obtenir en changeant uniquement sa propre stratégie.

Cette notion est utile car elle correspond mieux aux applications concrètes : dans de nombreux contextes, une très petite amélioration possible n'a pas d'importance réelle.

Cependant, il ne faut pas conclure que l'approximation rend toujours le problème facile. Tout dépend de la valeur de ε :

- si ε est relativement grand, on peut parfois trouver des équilibres approximés plus facilement ;
- si ε est très petit, par exemple inverse-polynomial en la taille du jeu, le problème reste difficile ;
- certains résultats montrent que calculer un équilibre approximé suffisamment précis reste PPAD-difficile.

Ainsi, l'approximation améliore souvent la situation en pratique, mais elle ne supprime pas complètement la difficulté théorique.

## 7. Comparaison entre équilibre exact et équilibre ε-approché

| Notion | Définition | Difficulté |
|---|---|---|
| Équilibre exact | Aucun joueur ne peut améliorer son gain en déviant seul | Très difficile dans le cas général ; lié à la PPAD-complétude |
| Équilibre ε-approché | Aucun joueur ne peut améliorer son gain de plus de ε | Plus flexible en pratique, mais reste difficile lorsque ε est petit |

La distinction est donc essentielle : l'équilibre exact correspond à la définition mathématique stricte, tandis que l'équilibre ε-approché autorise une marge d'erreur contrôlée. Cette marge peut rendre certains calculs plus réalistes, mais elle ne transforme pas nécessairement le problème en problème facile.

## Conclusion

Le problème du calcul d'un équilibre de Nash illustre bien la différence entre existence mathématique et calcul effectif. Le théorème de Nash garantit qu'un équilibre existe, mais la théorie de la complexité montre que le trouver peut être très difficile.

La classe **PPAD** permet de formaliser cette difficulté. Le résultat de PPAD-complétude de Daskalakis, Goldberg et Papadimitriou montre que le calcul d'un équilibre de Nash est l'un des problèmes centraux de cette classe. Enfin, la distinction entre équilibre exact et équilibre ε-approché est importante : l'approximation est souvent utile en pratique, mais le problème reste théoriquement difficile lorsque la précision demandée est élevée.
